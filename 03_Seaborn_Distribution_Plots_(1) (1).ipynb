# Distribution Plots

Let's discuss some plots that allow us to visualize the distribution of a data set. These plots are:

* distplot
* jointplot
* pairplot
* rugplot
* kdeplot

___
## Imports

```python
import seaborn as sns
```

## Data
Seaborn comes with built-in data sets!

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
tips.info()
```

## distplot

The distplot shows the distribution of a univariate set of observations.

```python
#count of every unique value in that column
#in displot on y axis value taken as density
sns.distplot(tips['size'],kde=True)
 
```

To remove the kde (kernel density estimate)  layer and just have the histogram use:

```python

# in his plot on y axis count value is taken
sns.histplot(tips['size'],kde=False)
```

## jointplot

jointplot() allows you to basically match up two distplots for bivariate data. With your choice of what **kind** parameter to compare with: 
* “scatter” 
* “reg” 
* “resid” 
* “kde” 
* “hex”

**it is uae to join two or more plots together**

```python
# it is uae to join two or more plots together
#kind is how data point is represent 
# scatter means reprsent using dot

sns.jointplot(x='total_bill',y='tip',data=tips,kind='scatter')
```

```python
# kind ='hex'  ---> means the data points represent using hexagoan design
# more density more dark hexagoan
sns.jointplot(x='total_bill',y='tip',data=tips,kind='hex')
```

```python
# reg means data represent throught the dots and draw regression line between them
# regression line shows the trend of data points
sns.jointplot(x='total_bill',y='tip',data=tips,kind='reg')
```

## pairplot

pairplot will plot pairwise relationships across an entire dataframe (for the numerical columns) and supports a color hue argument (for categorical columns). 

```python
# it is use to plot pairwise relationship of entire dataFrame
# if ralationship checks with the same colum's datapoints it return the histrogram
#represent the all columns data
sns.pairplot(tips)
```

```python
# hue -> is used to represent the data in term of the different color
#hue=sex
# means sex columns different data points represent by different color
#palette="coolwarm"
# in palette we choose own colors
# and coolwarm is theme of colors
sns.pairplot(tips,hue='sex',palette='coolwarm')
```

## rugplot

rugplots are actually a very simple concept, they just draw a dash mark for every point on a univariate distribution. They are the building block of a KDE plot:

```python
# it represent the data points by dash
sns.rugplot(tips['total_bill'])
```

## kdeplot

kdeplots are [Kernel Density Estimation plots](http://en.wikipedia.org/wiki/Kernel_density_estimation#Practical_estimation_of_the_bandwidth). These KDE plots replace every single observation with a Gaussian (Normal) distribution centered around that value. For example:

```python
#it is use to draw the kernel based on (dash,reg)data points 
sns.kdeplot(tips['total_bill'])
sns.rugplot(tips['total_bill']) 
```

```python
sns.kdeplot(tips['tip'])
sns.rugplot(tips['tip'])
```
